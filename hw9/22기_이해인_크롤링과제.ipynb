```python
import pandas as pd
import numpy as np
```

```python
import requests

url = 'https://www.yonsei.ac.kr/sc/support/calendar.jsp'
res = requests.get(url)

print(res)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.findAll('th')
```

```python
month = soup.find_all('th', {'scope':'row'})
```

```python
dayevent = soup.find_all('td',  attrs = {'class': "des"})
```

```python
df = pd.DataFrame()
monls = []

for i in range(len(month)):
    num = int(str(month[i]).split()[1].strip('rowspan="'))
    text = month[i].get_text()
    for n in range(num):
        monls.append(text)

df['Month'] = monls
df
```

```python
odd = range(0, len(dayevent), 2)
even = range(1, len(dayevent), 2)

dayls = []
for i in odd:
    text = dayevent[i].get_text().replace('\t', "")
    text = text.replace('\n', "")
    dayls.append(text)

df['Day'] = dayls

eventls = []
for i in even:
    text = dayevent[i].get_text()
    eventls.append(text)

df['Event'] = eventls

df
```
