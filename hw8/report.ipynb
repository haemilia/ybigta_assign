```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# data set: kaggle <Titanic - Machine Learning from Disaster>
# https://www.kaggle.com/competitions/titanic/overview
df = pd.read_csv('/content/train.csv')
df
```

```python
df_fam = df.copy()
df_fam['fam'] = df_fam['SibSp'] + df_fam['Parch']
df_fam['fam_01'] = df_fam.fam.apply(lambda x:1 if x >0 else 0 )
df_fam = df_fam[['Pclass', 'fam', 'fam_01']]
df_fam
```

```python
df_pie = df_fam.groupby('Pclass').count()
df_pie
```

```python
df_1 = df_fam[df_fam['Pclass'] == 1]
df_2 = df_fam[df_fam['Pclass'] == 2]
df_3 = df_fam[df_fam['Pclass'] == 3]
df_1 = df_1.groupby('fam_01').count()
df_2 = df_2.groupby('fam_01').count()
df_3 = df_3.groupby('fam_01').count()
```

```python
Pclass_labels = pd.Series(['1st', '2nd', '3rd'])
fam_labels = pd.Series(['Without Family', "With Family"])
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as po
import kaleido
```

```python
fig1 = px.pie(data_frame = df_pie, values = 'fam', names=Pclass_labels, title = 'Distribution of Passenger Class')
fig1.show()
fig1.write_image(file = "images/Pclass_pie.png")
```

```python
fig2 = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]])

fig2.add_trace(go.Pie(labels=fam_labels, values=df_1['fam'], name="1st Class"),
              1, 1)
fig2.add_trace(go.Pie(labels=fam_labels, values=df_2['fam'], name="2nd Class"),
              1, 2)
fig2.add_trace(go.Pie(labels=fam_labels, values=df_3['fam'], name="3rd Class"),
              1, 3)

fig2.update_traces(hoverinfo="label+percent+name")

fig2.update_layout(
    title_text="Each Passenger Class With or Without Family",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='1st', x=0.1, y=0.15, font_size=20, showarrow=False),
                 dict(text='2nd', x=0.5, y=0.15, font_size=20, showarrow=False),
                 dict(text='3rd', x=0.9, y=0.15, font_size=20, showarrow=False)])

fig2.show()
fig2.write_image(file = "images/WithWithoutPie.png")
```

```python
df_bar = df_fam.groupby(['Pclass', 'fam'])['fam'].count().to_frame(name = 'count').reset_index()
bardict = {1: '1st', 2: '2nd', 3: '3rd'}
df_bar = df_bar.replace({"Pclass":bardict})
```

```python
label = {"Pclass": "Passenger Class", "fam": "# of Family Member"}
fig3 = px.bar(df_bar, x = 'Pclass', y = 'count', color = 'fam', labels= label)

fig3.show()
fig3.write_image(file = "images/FamilyMemberBar.png")
```
